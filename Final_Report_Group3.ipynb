# **Air quality in different district in 2022(Heatmap)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the provided CSV file
data = pd.read_csv('air.csv')

# Extract the data for the year 2022
data_2022 = data[data['YEAR'] == 2022]

# Define the stations and pollutants
stations = data_2022['STATION'].unique()
pollutants = ['SO2', 'NOX', 'NO2', 'CO', 'RSP', 'O3', 'FSP']

# Create a matrix to store the pollutant amounts
matrix = np.zeros((len(pollutants), len(stations)))

# Fill the matrix with pollutant amounts
for i, pollutant in enumerate(pollutants):
    for j, station in enumerate(stations):
        value = data_2022.loc[(data_2022['STATION'] == station), pollutant].values
        if len(value) > 0 and value[0] != 'N.A.':
            matrix[i, j] = float(value[0])

# Plot the heatmap
fig, ax = plt.subplots(figsize=(10, 6))
im = ax.imshow(matrix, cmap='YlOrBr')

# Customize the plot
ax.set_xticks(np.arange(len(stations)))
ax.set_yticks(np.arange(len(pollutants)))
ax.set_xticklabels(stations)
ax.set_yticklabels(pollutants)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Add the pollutant values to each box
for i in range(len(pollutants)):
    for j in range(len(stations)):
        text = ax.text(j, i, int(matrix[i, j]), ha="center", va="center", color="black")

# Add a color bar
cbar = ax.figure.colorbar(im, ax=ax, format="%d")
cbar.ax.set_ylabel("Pollutant Amount", rotation=-90, va="bottom")

# Set the title and show the plot
ax.set_title("Air Quality Heatmap - 2022")
plt.tight_layout()
plt.show()
```

## **Result**


*   The air quality in Causeway Bay, Central and Mongkok is worse than in other areas. (Has the highest amount of most pollutants, lowest amount of Ozone)
*   Ozone levels can be influenced by several factors including sunlight, temperature, and the presence of other pollutants. In urban area like Causeway Bay, Central and Mong Kok, buildings and other structures can block sunlight, reducing the amount of ozone that is produced.

# **Scatterplot with linear regression**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Read the data from the provided CSV file
data = pd.read_csv('air.csv')

# Filter the data for the years 2012 to 2022
data_filtered = data[(data['YEAR'] >= 2012) & (data['YEAR'] <= 2022)]

# Convert NOx and ozone columns to numeric data types
data_filtered['NOX'] = pd.to_numeric(data_filtered['NOX'], errors='coerce')
data_filtered['O3'] = pd.to_numeric(data_filtered['O3'], errors='coerce')

# Remove rows with NaN values in NOx and ozone columns
data_filtered = data_filtered.dropna(subset=['NOX', 'O3'])

# Select NOx and ozone columns
nox = data_filtered['NOX']
ozone = data_filtered['O3']

# Create a scatterplot
plt.scatter(nox, ozone, alpha=0.5)

# Set plot title and labels
plt.title('Scatterplot of NOx and Ozone (2012-2022)')
plt.xlabel('NOx')
plt.ylabel('Ozone')

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(nox, ozone)
line = slope * nox + intercept

# Add linear regression line to the plot
plt.plot(nox, line, color='red')

# Display the plot
plt.show()
```

## **Result**

From the graph, we conclude that the reduction of NOx emissions can actually lead to an increase in Ozone level. Here is the reason.


*   In area with high NOx emissions, like urban environments, the excess Nox can help to remove Ozone from the atmosphere through a process called titration
*   When NOx levels are reduced, this titration process is less effective, and Ozone levels can increase

# **Hypothesis testing(Is there a significant difference in NOx emission between different seasons in Hong Kong?)**

```python
import pandas as pd
from scipy.stats import f_oneway

# Read the data from the CSV file
df = pd.read_csv('a.csv')

# Extract the data for each season (assuming seasons: Winter, Spring, Summer, Autumn)
winter_data = df[['January', 'February', 'December']]
spring_data = df[['March', 'April', 'May']]
summer_data = df[['June', 'July', 'August']]
autumn_data = df[['September', 'October', 'November']]

# Perform one-way ANOVA test for NOx emission
f_statistic, p_value = f_oneway(winter_data.values.flatten(),
                                spring_data.values.flatten(),
                                summer_data.values.flatten(),
                                autumn_data.values.flatten())

# Print the results
print("One-way ANOVA results for NOx emission:")
print("F-statistic:", f_statistic)
print("p-value:", p_value)

# Check if the p-value is below the significance level (e.g., 0.05) to reject the null hypothesis
if p_value < 0.05:
    print("There is a significant difference in NOx emission between different seasons in Hong Kong.")
else:
    print("There is no significant difference in NOx emission between different seasons in Hong Kong.")
```


## **Result**

**Null hypothesis (H0)**: There is no significant difference in NOx emission between different seasons in Hong Kong.

**Alternative hypothesis (HA)**: There is a significant difference in NOx emission between different seasons in Hong Kong.

The p-value is lower than 0.05. It shows that there is a significant difference in NOx emission between different seasons in Hong Kong.





### **Mathematical formulas**
**F-statistic:** The F-statistic is a measure of the variation between group means compared to the variation within groups. A higher F-statistic value suggests a larger difference between the group means.

**p-value:** The p-value represents the probability of observing the obtained F-statistic (or a more extreme value) under the null hypothesis. A p-value below the significance level (e.g., 0.05) indicates that the difference in NOx emission between seasons is statistically significant.

p-value = P(F > F_observed | H0)

F as the calculated F-statistic.

F_observed represents the observed F-statistic obtained from the one-way ANOVA test.

# **Geospatial Visualization(Hong Kong air quality in 2022)**




```python
import pandas as pd
import folium

data = pd.read_csv('airg.csv')

data_2022 = data[data['YEAR'] == 2022]

stations = ['CAUSEWAY BAY', 'CENTRAL', 'CENTRAL/WESTERN', 'EASTERN', 'KWAI CHUNG', 'KWUN TONG',
            'MONG KOK', 'SHAM SHUI PO', 'SHATIN', 'TAI PO', 'TAP MUN', 'TSUEN WAN', 'TUNG CHUNG',
            'YUEN LONG', 'TUEN MUN', 'TSEUNG KWAN O', 'YUEN LONG', 'TSEUNG KWAN O']

pollutants = ['SO2', 'NOX', 'CO','NO2','RSP','O3','FSP']  # List of pollutants for control menu

station_locations = {
    'CAUSEWAY BAY': [22.2795, 114.1827],
    'CENTRAL': [22.2829, 114.1589],
    'CENTRAL/WESTERN': [22.2868, 114.1494],
    'EASTERN': [22.2731, 114.2360],
    'KWAI CHUNG': [22.3707, 114.1234],
    'KWUN TONG': [22.3129, 114.2222],
    'MONG KOK': [22.3204, 114.1698],
    'SHAM SHUI PO': [22.3317, 114.1595],
    'SHATIN': [22.3770, 114.1974],
    'TAI PO': [22.4504, 114.1655],
    'TAP MUN': [22.3696, 114.3429],
    'TSUEN WAN': [22.3724, 114.1133],
    'TUNG CHUNG': [22.2871, 113.9428],
    'YUEN LONG': [22.4445, 114.0222],
    'TUEN MUN': [22.3959, 113.9777],
    'TSEUNG KWAN O': [22.3070, 114.2565]
}

hongkong_map = folium.Map(location=[22.3193, 114.1694], zoom_start=11)

# Create feature groups for each pollutant
feature_groups = {pollutant: folium.FeatureGroup(name=f"{pollutant} Emissions") for pollutant in pollutants}

for pollutant in pollutants:
    pollutant_data = data_2022[data_2022['STATION'].isin(stations)][['STATION', pollutant]]

    for index, row in pollutant_data.iterrows():
        station = row['STATION']
        pollutant_value = row[pollutant]
        if station in station_locations:
            latitude, longitude = station_locations[station]
            popup_text = f"Station: {station}<br>{pollutant} Emission: {pollutant_value} µg/m³"
            marker = folium.Marker(location=[latitude, longitude], popup=popup_text)
            marker.add_to(feature_groups[pollutant])  # Add marker to corresponding feature group

# Add feature groups to the map
for feature_group in feature_groups.values():
    feature_group.add_to(hongkong_map)

folium.LayerControl().add_to(hongkong_map)  # Layer control menu

hongkong_map
```

## **Result**
The map can show different pollutant emissions in 2022 by using the filter to choose which pollutant to be displayed. Clicking the maker shows the detailed data.
